### 4.Data Cleaning

```python
data.isnull().sum()
```

```python
null_counts = data.isnull().sum()
(null_counts/len(data))*100
```

```python
##Dropping Null records since the percentage of NULL values is very less 
data.dropna(inplace=True)

data.isnull().sum()

data.duplicated().sum()

data.drop_duplicates(inplace= True)

data[data.duplicated()]
```

### 5.Data Analysis 

**Univariant Analysis**


```python
## Displaying the different neighbour_hoods with a bar graph.

neighbours1 = data.groupby('neighbourhood')[['neighbourhood']].value_counts().reset_index(name = 'count')
neighbours1
```

```python
from matplotlib import pyplot as plt


sns.countplot(data = data,x = 'neighbourhood_group')
plt.title("No of Airbnb on different neighbourhoods")
plt.xlabel("Neighbour")
plt.show()
```

```python
## Price distribution across the dataset 

##filtering outliers 

df = data[data['price'] <= 1500]

sns.boxplot(data = df,x='price')

```

```python
px.figure(figsize=(8,4))
sns.histplot(df,x='price',bins= 50)
px.ylabel('Frequency')
px.title('Price Distrution')
```

```python
## Checking availablity of the airbnb with 365

px.figure(figsize=(8,4))
sns.histplot(data,x='availability_365',bins= 50)
px.ylabel('Frequency')
px.title('Availability Distrution')
```

```python
## Avg price in each neighbour hood 

avg_price_neighbour = data.groupby(by = 'neighbourhood_group')[['price']].mean()
##avg_price_neighbour = avg_price_neighbour.to_frame()

no_of_units = data.groupby(by = 'neighbourhood_group')['neighbourhood_group'].count()
no_of_units = no_of_units.to_frame()

##pd.concat([avg_price_neighbour,no_of_units],ignore_index=True)
```

```python
##Feature Enginerring 

##avg price per bed 

##adding a new column price by bed

data['price_per_bed'] = data['price']/data['beds']
data.head()
```

```python
neighbour_per_bed_avg = data.groupby('neighbourhood_group')['price_per_bed'].mean()
neighbour_per_bed_avg
```

```python
## Bi vairant analysis 
##how one column effects another column 

sns.barplot(data = data, x = 'neighbourhood_group',y='price',hue='room_type')
```
